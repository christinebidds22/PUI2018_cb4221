```python
# renaming columns
jan_cabs_trimmed=jan_cabs_trimmed.rename(columns = {'tpep_dropoff_datetime' : 'dropoff_time'})
jan_cabs_trimmed=jan_cabs_trimmed.rename(columns = {'PULocationID' : 'pickup_zone'})
jan_cabs_trimmed=jan_cabs_trimmed.rename(columns = {'DOLocationID' : 'dropoff_zone'})
jan_cabs_trimmed.head()                                       
```

```python
# selecting columns I wanted to remaini in df
jan_cabs_trimmed2=jan_cabs_trimmed[['tpep_pickup_datetime', 'dropoff_time', 'passenger_count', 'trip_distance', 'pickup_zone',
                                  'dropoff_zone', 'payment_type', 'fare_amount', 'total_amount']]
jan_cabs_trimmed2.head()
```

```python
figure = jan_cabs_trimmed2.plot(x='trip_distance', y='total_amount', kind='scatter')
pl.xlabel("Trip Distance")
pl.ylabel("Total Fare")
pl.title("Influence of Trip Distance on Taxi Fares")
pl.legend()
```

**Figure 1:** Total Fare vs. Trip Distance. As you might expect, there is a strong positive relationship between trip distance and total fare - the longer the trip, the higher the price. 

## IMPROVING MY PLOT BASED ON FEEDBACK

- Fitting a line
- Adding units to axes
- Zooming in to see the trend better

```python
# fitting a line 

model = smf.ols(formula='total_amount ~ trip_distance', data=jan_cabs_trimmed2).fit()
model.summary()
```

```python
x = jan_cabs_trimmed2['trip_distance']
jan_cabs_trimmed2.plot(x='trip_distance', y='total_amount', kind='scatter', figsize=(10,10))
pl.plot(x, model.predict(jan_cabs_trimmed2), '-', c='r', label="fitted regression, R2=.869")
pl.xlabel("Trip Distance (miles)", fontsize=16)
pl.ylabel("Total Fare ($US)", fontsize=16)
pl.xlim(1, 33)
pl.ylim(1, 125)
pl.title("Influence of Trip Distance on Taxi Fares", fontsize=20)
pl.legend()
```

Figure 2: Data from the NYC Taxi and Limousine Commission from January 2018. There is a reasonably clear linear relationship between trip distance and total fare, and the fitted line has an R^2 value of .869. Things get more inconsistent for trips further than 20 miles or more than 60 dollars, perhaps because they are not many of them. The majority of the data are clustered in the bottom left corner of the view, i.e. trips of 10 miles or less and fares 40 dollars or less.
